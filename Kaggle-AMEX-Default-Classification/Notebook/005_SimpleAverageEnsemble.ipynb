```python
import pandas as pd
from glob import glob
```

```python
outputs = glob('*.csv')
IS_FIRST = True

while outputs:
    if IS_FIRST:
        df = pd.read_csv(outputs.pop(0))
        IS_FIRST = False
    df['prediction'] = (df['prediction'].values + pd.read_csv(outputs.pop(0))['prediction'].values) / 2
```

```python
df.to_csv('999_AverageEnsemble.csv', index=False)
```
